```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
plt.rc("font",family="Malgun Gothic")
```

```python
df_raw= pd.read_csv("Downloads/국가_대륙_별_상품군별_온라인쇼핑_해외직접판매액_20201009135308.csv",encoding="cp949",engine="python")
df_raw.head()
```

```python
df_raw["국가(대륙)별"].value_counts()
```

```python
df_raw[df_raw["국가(대륙)별"]=="미국"]
```

```python
# melt를 통해 column에서 값으로
df=df_raw.melt(id_vars=["국가(대륙)별","상품군별","판매유형별"], var_name="기간",value_name="백만원")
df.head()
```

```python
df.info()
```

```python
# split 후 0번째 인덱스
int("2019 4/4 p".split()[0])
```

```python
# 년도뽑기
# lambda = 함수 지정
df["연도"]=df["기간"].map(lambda x: int(x.split()[0]))
```

```python
int("2019 4/4 p".split()[1].split("/")[0])
```

```python
df["분기"]=df["기간"].map(lambda x: int(x.split()[1].split("/")[0]))
```

```python
df.head()
```

```python
df["백만원"]=df["백만원"].replace("-",np.nan).astype(float)
df.head()
```

```python
# != --> 필요한 데이터 남기고 제외
df= df[(df["국가(대륙)별"]!="합계") & (df["상품군별"]!="합계")].copy()
```

```python
df.isnull().sum()
```

```python
df_total=df[df["판매유형별"]=="계"].copy()
df_total.head()
```

```python
sns.lineplot(data=df_total,x="연도",y="백만원")
```

```python
#legend => 목록값 
sns.lineplot(data=df_total,x="연도",y="백만원",hue="상품군별")
```

```python
# relplot --> 서브플롯으로 더 자세하게 분석가능
# kind = 차트종류, col= 세분화할 컬럼, col_wrap= 한줄에 몇개
sns.relplot(data=df_total, x="연도",y="백만원",hue="상품군별",kind="line", col="상품군별", col_wrap=4)
```

```python
# ~:조건 아닌 경우 
df_sub=df_total[~df_total["상품군별"].isin(["화장품","의류 및 패션관련 상품"])].copy()
```

```python
sns.relplot(data=df_sub, x="연도",y="백만원", kind="line", hue="상품군별", col="상품군별", col_wrap=4 )
```

```python
df_cos=df_total[df_total["상품군별"]=="화장품"].copy()
df_cos.head()
```

```python
plt.figure(figsize=(15,4))
sns.lineplot(data=df_cos,x="연도",y="백만원",hue="분기")
```

```python
plt.figure(figsize=(18,4))
sns.lineplot(data=df_cos,x="기간",y="백만원", hue="국가(대륙)별")
```

```python
plt.figure(figsize=(18,4))
# x축이 빽빽할경우
plt.xticks(rotation=30)
# 중국 빼고
sns.lineplot(data=df_cos[df_cos["국가(대륙)별"] != "중국"],x="기간",y="백만원", hue="국가(대륙)별")
```

```python
# != : 조건인 경우
plt.figure(figsize=(18,4))
plt.xticks(rotation=30)
df_sum=df[(df["판매유형별"] !="계")&(df["판매유형별"]!="면세점")].copy()
sns.lineplot(data=df_sum, x="기간",y="백만원",hue="판매유형별",ci=None)
```

```python
df_fashion=df[(df["상품군별"]== "의류 및 패션관련 상품")&(df["판매유형별"]=="계")].copy()
df_fashion.head()
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion,x="기간",y="백만원",hue="판매유형별",ci=None)
```

```python
result = df_fashion.pivot_table(index="국가(대륙)별",columns="연도",
                       values="백만원",aggfunc="sum")
result
```

```python
# annot = 숫자 보이게, fmt= 소수 몇째자리까지
sns.heatmap(result, cmap="Blues",annot=True, fmt=".0f")
```
